```python
data['Description'] = data['Description'].str.strip()
```

```python
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
```

```python
import random
random.seed(3)
negara=random.sample(list(data["Country"].unique()),15)
negara
```

```python
df=data.loc[(data['Country'] == 'Iceland') | (data['Country'] == 'USA')|(data['Country'] == 'Spain')|(data['Country'] == 'Bahrain')|(data['Country'] == 'Czech Republic')|(data['Country'] == 'Finland')|(data['Country'] == 'EIRE')|(data['Country'] == 'Switzerland')|(data['Country'] == 'Denmark')|(data['Country'] == 'European Community')|(data['Country'] == 'RSA')|(data['Country'] == 'Singapore')|(data['Country'] == 'Saudi Arabia')|(data['Country'] == 'United Kingdom')|(data['Country'] == 'Germany')]
df
```

```python
basket = (df.groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
basket
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.03, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
